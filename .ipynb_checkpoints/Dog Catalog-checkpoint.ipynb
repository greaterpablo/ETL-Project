```python
#Dependencies
from bs4 import BeautifulSoup
import requests
from splinter import Browser
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
```

 ## American Kennel Club Scraping

```python
#Visiting American Kennel Club website-Dog Brees
url = 'https://www.akc.org/dog-breeds/'

# Setup splinter
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

Scrapping the data from url mentioned before (American Kennel Club) to retrieve the official dog breeds

```python
browser.visit(url)
html = browser.html
```

```python
soup = BeautifulSoup(html, 'html.parser')
```

```python
#Letter obtainining from main page
links = soup.find_all('div',class_="breed-letters-filter__letter")
```

```python
#constructing link to scrape the breed for each letter.
url_2 = 'https://www.akc.org'
links_to_visit = [] 
for link in links:
    try:
        links_to_visit.append(url_2+link.a['href'])
        print(link.a['href'])
    except:
        print("No link for letter found")
```

```python
#Getting the breeds by letter
list_breeds = []
list_breeds_name = []
for browse in links_to_visit:
    browser.visit(browse)
    html = browser.html
    soup = BeautifulSoup(html, 'html.parser')
    all_species = soup.find_all("div",class_="breed-type-card")
    for specie in all_species:
        list_breeds_name.append(specie["data-title"])
        list_breeds.append({"Breed":specie["data-title"],"Short Description":specie.p.text})
        print(specie["data-title"])
```

```python
list_breeds
```

## Dogtime Scraping

Scrapping from dogtime.com the most important information for each dog

```python
url = "https://dogtime.com/dog-breeds/profiles"
browser.visit(url)
html = browser.html
soup = BeautifulSoup(html,'html.parser')
```

```python
dog_links = soup.find_all("a", class_ = "list-item-title")
```

```python
##All the dogs available, but there are mixed breeds
for dog in dog_links:
        print(dog.text)
```

```python
##Looping all records from the list_dictionary into the dogs available in dogtime and getting rating for certain characteristics
for dictio in list_breeds:
    for dog in dog_links:
        if dog.text == dictio['Breed']:
            chars = []
            stats = []
            browser.visit((dog["href"]))
            html = browser.html
            soup = BeautifulSoup(html,'html.parser')
            characteristic = soup.find_all("h3", class_="characteristic-title")
            for char in characteristic:
                chars.append(char.text)
            ratings = soup.find_all("div", class_="characteristic-stars parent-characteristic")
            for div in ratings:
                r = div.find("div", class_ = "characteristic-star-block")
                stats.append(r.div['class'][1].split('-')[1])
            for ins in range(len(chars)):
                dictio[chars[ins]] = stats[ins]
        
            
            
            
    
```

## CSV Information Extraction

```python
best_pd = pd.read_csv("Resources/best_in_show.csv")
df_best = pd.DataFrame(best_pd)
df_best.head()
```

```python
df_best[["LIFETIME COST, $","2 LONGEVITY","3 NO. OF GENETIC AILMENTS","4a average purchase price, US$","intelligence category"]]
df_best
```

```python
final_best = df_best[["Dog breed","LIFETIME COST, $","2 LONGEVITY","3 NO. OF GENETIC AILMENTS","4a average purchase price, US$","intelligence category"]]
final_best
```

```python
final_best = final_best.rename(columns = {"Dog breed" : "Breed", "LIFETIME COST, $":"Lifetime Cost","2 LONGEVITY":"Longevity","3 NO. OF GENETIC AILMENTS": "Genetic Ailments","4a average purchase price, US$":"Purchase Price", "intelligence category":"Intelligence Category"},inplace = False)
final_best = final_best.iloc[1:]
final_best = final_best.set_index("Breed")
final_best
```

```python
size = pd.read_csv("Resources/Dog_weight_length.csv", encoding = "latin")
df_size = pd.DataFrame(size)
df_size = df_size.set_index("Breed")
df_size
```

```python
characteristics = pd.read_csv("Resources/dog_breed_characteristics.csv")
df_characteristics = pd.DataFrame(characteristics)
df_characteristics
```

```python
df_characteristics = df_characteristics.rename(columns = {"BreedName":"Breed"},inplace = False)
df_characteristics = df_characteristics.iloc[1:]
df_characteristics
```

```python
df_characteristics = df_characteristics[["Breed","Temperment","Watchdog"]]
df_characteristics = df_characteristics.replace(['Xoloitzcuintle/Mexican Hairless'],'Xoloitzcuintle')
df_characteristics = df_characteristics.set_index("Breed")
df_characteristics
```

```python
for dictio in list_breeds:
    if dictio["Breed"] in final_best.index:
        try:
            dictio["Lifetime Cost"] = final_best.loc[final_best.index == dictio["Breed"]]["Lifetime Cost"][0]
            dictio["Longevity"] = final_best.loc[final_best.index == dictio["Breed"]]["Longevity"][0]
            dictio["Genetic Ailments"]  = final_best.loc[final_best.index == dictio["Breed"]]["Genetic Ailments"][0]
            dictio["Purchase Price"] = final_best.loc[final_best.index == dictio["Breed"]]["Purchase Price"][0]
            dictio["Intelligence Category"] = final_best.loc[final_best.index == dictio["Breed"]]["Intelligence Category"][0]
        except:
            print("An exception occured in "+dictio["Breed"])
    if dictio["Breed"] in df_size.index:
        try:
            dictio["Avg Size"] = (int(df_size.loc[df_size.index == dictio["Breed"]]["weight_low_lbs"][0]) + int(df_size.loc[df_size.index == dictio["Breed"]]["height_high_inches"][0]))/2
            dictio["Avg Weight"] = (int(df_size.loc[df_size.index == dictio["Breed"]]["weight_low_lbs"][0]) + int(df_size.loc[df_size.index == dictio["Breed"]]["weight_high_lbs"][0]))/2                   
        except:
            print("An exception occured in "+dictio["Breed"])
    if dictio["Breed"] in df_characteristics.index:
        try:
            dictio["Temperment"] = df_characteristics.loc[df_characteristics.index == dictio["Breed"]]["Temperment"][0]
            dictio["Watchdog Rating"] = df_characteristics.loc[df_characteristics.index == dictio["Breed"]]["Watchdog"][0]
        except:
            print("An exception occured in "+dictio["Breed"])
```
