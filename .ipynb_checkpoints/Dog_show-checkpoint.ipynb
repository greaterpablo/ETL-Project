# Best dogs in show ( UK, USA)

```python
import pandas as pd 
```

Importing winning dog tables from Wikipedia from the National Dog Show and Westminster Dog Show 

```python
#Westminster URL
w_url = 'https://en.wikipedia.org/wiki/List_of_Best_in_Show_winners_of_the_Westminster_Kennel_Club_Dog_Show'
```

```python
#National show URL 
n_url = "https://en.wikipedia.org/wiki/National_Dog_Show"
```

### Westminster dog show 

```python
# Use Panda's `read_html` to parse the url
### BEGIN SOLUTION
tables = pd.read_html(w_url)
tables
```

```python
df = tables[1]
df.columns = ['year', 'winner', 'image', 'breed', 'group', 'owner', 'ref(s)']
df.head()
```

```python
W_dogs= df.filter(['year','winner','breed'])
W_dogs.head()
```

### National Dog Show

```python
W_dogs.value_counts("breed")
```

```python
# Use Panda's `read_html` to parse the url
### BEGIN SOLUTION
tables = pd.read_html(n_url)
tables
```

```python
df_2 = tables[1]
df_2.columns = ['year', 'breed', 'name', 'call name', 'owner/trainer', 'reference']
df_2.head()
```

```python
N_dogs= df_2.filter(['year','name','breed'])
N_dogs.head()
```

```python
N_dogs.rename(columns ={"name":"winner"})
```

```python
N_dogs.value_counts('breed')
```
